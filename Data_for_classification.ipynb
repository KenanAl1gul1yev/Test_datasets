```python
import pandas as pd
import numpy as np
df = pd.read_csv('Data_for_classification.csv')
df
```

```python
cols_to_drop = ['fnlwgt', 'race']
df.drop(cols_to_drop, axis = 'columns', inplace = True)
df
```

```python
cols = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'sex', 'native-country']
for i in cols:
    print(df.groupby(i)[i].agg('count').sort_values(ascending = False))
```

```python
df['income'].unique()
```

```python
df.replace({'?': np.NaN, '>50K.': '>50K', '<=50K.': '<=50K'}, inplace = True)
df
```

```python
df['income'].unique()
```

```python
df.dropna(inplace = True)
df
```

```python
numeric_cols = ['age', 'education-num', 'capital-gain', 'capital-loss','hours-per-week']
categorical_cols = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'sex', 'native-country', 'income']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
df
```

```python
df1 = pd.get_dummies(data = df, columns = categorical_cols, drop_first = True)
df1
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df1.drop(['income_>50K'], axis = 'columns'), df1['income_>50K'], test_size = 0.2, random_state = 42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 100)
clf.fit(x_train, y_train)
clf.score(x_test, y_test)
```
